```python
filenames = ['angry','fearful','happy','neutral','sad','surprised']
```

```python
def read_dataset(filename):
  path_1 = 'gdrive/Shareddrives/Deep Learning Group Work/archive/train/'+filename
  dir_list_1 = os.listdir(path_1)
  df_1 = np.zeros((len(dir_list_1),42,42))
  index = 0
  flag = 0
  for i in dir_list_1:
      if i != '.DS_Store':
          Im_path = path_1 + '/' + i
          img = iio.imread(Im_path)
          img_pil = Image.fromarray(img)
          img_42_42 = np.array(img_pil.resize((42, 42), Image.ANTIALIAS))
          df_1[index] = img_42_42.astype(int)
          index += 1
          if index == len(dir_list_1):
              flag = 1
              break
      if flag == 1:
          break           
      path = 'gdrive/Shareddrives/Deep Learning Group Work/archive/train/'+filename
  df_1 = df_1.astype(int) 
  return df_1
```

```python
dfs = []
for filename in filenames:
  df_temp = read_dataset(filename)
  dfs.append(df_temp)
```

```python
import random

guess = 0
attempts = 0

print ("Guess the Emotion!")
print ("=================================")
print ("Instructions:")
print ("Guess the emotion based off the image!")
print ("Input 0 for angry")
print ("Input 1 for fearful")
print ("Input 2 for happy")
print ("Input 3 for neutral")
print ("Input 4 for sad")
print ("Input 5 for surprised")


inputs = []
answers = []
index = []
n = 10

while attempts < n:
    
    emotion = random.randint(0,5)
    img = random.randint(0,300)
    attempts += 1
    plt.imshow(dfs[emotion][img])
    plt.title(attempts)
    plt.show()
    answers.append(emotion)

    guess = input("Guess emotion: ")
    guess = int(guess)
    inputs.append(guess)
    index.append(attempts)

print("Index: ")   
print(index)
print("Your guess: ") 
print(inputs)  
print("Actual answers: ")
print(answers)               
accuracy = round(sum(1 for x,y in zip(inputs,answers) if x == y) / len(answers) * 100,2)
print ("Accuracy: " + str(accuracy) + "%")
if (accuracy < 34):
  print("Oof! Better luck next time :(")
elif(accuracy < 67):
  print("Nice Job! Your EQ is solid :)")
else:
  print("Wow! Your brain is gigantic! :D")
```
